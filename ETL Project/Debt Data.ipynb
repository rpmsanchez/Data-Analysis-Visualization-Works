```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
debt_file = "Resources/debt.csv"
debtdf = pd.read_csv(debt_file)

pov_file = "Resources/poverty_headcount.csv"
povdf = pd.read_csv(pov_file)

rev_file = "Resources/revenue.csv"
revdf = pd.read_csv(rev_file)

WGI_file = "Resources/WGI_Data.csv"
wgidf = pd.read_csv(WGI_file)
```

```python
debtdf = debtdf.iloc[:217]

povdf = povdf.iloc[:217]

revdf = revdf.iloc[:217]
```

```python
debtdf = debtdf.replace('..', 'NaN')

povdf = povdf.replace('..', 'NaN')

revdf = revdf.replace('..', 'NaN')
```

```python
#DEBT DATA
debtdf = debtdf.rename(columns={"2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017", "2018 [YR2018]":"2018"})

debtdf = debtdf.drop(columns=['Series Name', 'Series Code', '1990 [YR1990]', '2000 [YR2000]'])
debtdf.head()

#POVERTY DATA
povdf = povdf.rename(columns={"2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017", "2018 [YR2018]":"2018"})

povdf = povdf.drop(columns=['Series Name', 'Series Code', '1990 [YR1990]', '2000 [YR2000]'])
povdf.head()

#REVENUE DATA
revdf = revdf.rename(columns={"2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017", "2018 [YR2018]":"2018"})

revdf = revdf.drop(columns=['Series Name', 'Series Code', '1990 [YR1990]', '2000 [YR2000]'])
revdf.head()
```

```python
debtdf = debtdf.melt(id_vars=['Country Name', 'Country Code'], value_vars=['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']\
                     , var_name ='Year', value_name ='Central government debt, total (% of GDP)')
debtdf

povdf = povdf.melt(id_vars=['Country Name', 'Country Code'], value_vars=['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']\
                     , var_name ='Year', value_name ='Poverty headcount ratio at $1.90 a day (% of pop)')
povdf

revdf = revdf.melt(id_vars=['Country Name', 'Country Code'], value_vars=['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']\
                     , var_name ='Year', value_name ='Revenue, excluding grants (% of GDP)')
revdf
```

```python
worldbank_df = debtdf.merge(povdf, on=('Country Name', 'Country Code', 'Year'))\
                       .merge(revdf, on=('Country Name', 'Country Code', 'Year'))
pd.set_option('display.max_rows', 500)
worldbank_df = worldbank_df.astype({'Year': float, "Central government debt, total (% of GDP)": float, \
                                    "Poverty headcount ratio at $1.90 a day (% of pop)": float, \
                                   'Revenue, excluding grants (% of GDP)': float})
worldbank_df.dtypes
```

```python
connection_string = "root:getserved101@localhost/worldbank_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
engine.table_names()
```

```python
worldbank_df.to_sql(name='worldbank', con=engine, if_exists='append', index=False)

```
